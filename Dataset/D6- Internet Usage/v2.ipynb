```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.cm as cm
import matplotlib.colors as mcolors
import numpy as np
import pandas as pd

# Stile dark coerente
plt.style.use("dark_background")
sns.set_style("white")
plt.rcParams.update({
    "figure.facecolor": "#050b18",
    "axes.facecolor": "#050b18",
    "savefig.facecolor": "#050b18",
    "axes.edgecolor": "#cccccc",
    "axes.labelcolor": "#cccccc",
    "xtick.color": "#cccccc",
    "ytick.color": "#cccccc",
    "text.color": "#cccccc",
    "font.family": "monospace",
    "font.weight": "bold"
})

# Carica dataset
df = pd.read_csv("Final.csv")

# 1. Crescita globale degli utenti Internet
def plot_total_users():
    global_data = df.groupby("Year")["No. of Internet Users"].sum().reset_index()
    plt.figure(figsize=(10, 5))
    plt.plot(global_data["Year"], global_data["No. of Internet Users"] / 1e9, marker="o", linewidth=2)
    plt.title("Crescita globale degli utenti Internet (miliardi)", fontweight="bold")
    plt.xlabel("Anno", fontweight="bold")
    plt.ylabel("Utenti Internet (miliardi)", fontweight="bold")
    plt.tick_params(axis='both', labelsize=10)
    plt.show()

# 2. Percentuale media globale di utenti Internet
def plot_percentage_users():
    avg_pct = df.groupby("Year")["Internet Users(%)"].mean().reset_index()
    plt.figure(figsize=(10, 5))
    plt.plot(avg_pct["Year"], avg_pct["Internet Users(%)"], marker="o", color="deepskyblue", linewidth=2)
    plt.title("Percentuale media globale di utenti Internet", fontweight="bold")
    plt.xlabel("Anno", fontweight="bold")
    plt.ylabel("Percentuale (%)", fontweight="bold")
    plt.tick_params(axis='both', labelsize=10)
    plt.show()

# 5. Top 10 paesi per numero di utenti Internet
def plot_top10_users():
    df_latest = df[df["Year"] == df["Year"].max()]
    df_filtered = df_latest[df_latest["Entity"] != "World"]
    top10 = df_filtered.sort_values("No. of Internet Users", ascending=False).head(10)

    plt.figure(figsize=(10, 6))
    sns.barplot(data=top10, x="No. of Internet Users", y="Entity", palette="viridis")
    plt.title("Top 10 paesi per numero di utenti Internet (2023)", fontweight="bold")
    plt.xlabel("Utenti Internet", fontweight="bold")
    plt.ylabel("Paese", fontweight="bold")
    plt.tick_params(axis='both', labelsize=10)
    plt.show()

# 9. Top 20 paesi per % utenti Internet con gradiente ordinato
def plot_pct_top20_ordered_gradient():
    df_latest = df[df["Year"] == df["Year"].max()]
    df_2023 = df_latest[df_latest["Internet Users(%)"] > 0].copy()
    top20 = df_2023.sort_values("Internet Users(%)", ascending=False).head(20)
    top20 = top20.sort_values("Internet Users(%)", ascending=True)

    # Gradiente basato sulla posizione (non sul valore)
    cmap = cm.RdYlGn
    colors = cmap(np.linspace(0, 1, len(top20)))

    plt.figure(figsize=(10, 8))
    plt.barh(top20["Entity"], top20["Internet Users(%)"], color=colors)
    plt.xlabel("% Popolazione con accesso a Internet (2023)", fontweight="bold")
    plt.title("Top 20 paesi per % utenti Internet (gradiente ordinato)", fontweight="bold")
    plt.tick_params(axis='both', labelsize=10)
    plt.show()

# Esecuzione
plot_total_users()
plot_percentage_users()
plot_top10_users()
plot_pct_top20_ordered_gradient()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.cm as cm
import numpy as np
import pandas as pd

# Stile coerente con presentazione dark
plt.style.use("dark_background")
sns.set_style("darkgrid")
plt.rcParams["figure.facecolor"] = "#050b18"
plt.rcParams["axes.facecolor"] = "#050b18"
plt.rcParams["savefig.facecolor"] = "#050b18"
plt.rcParams["axes.edgecolor"] = "#cccccc"
plt.rcParams["axes.labelcolor"] = "#cccccc"
plt.rcParams["xtick.color"] = "#cccccc"
plt.rcParams["ytick.color"] = "#cccccc"
plt.rcParams["text.color"] = "#cccccc"
plt.rcParams["font.family"] = "monospace"

# Carica dati
df = pd.read_csv("Final.csv")

# 1. Crescita globale degli utenti Internet
def plot_total_users():
    global_data = df.groupby("Year")["No. of Internet Users"].sum().reset_index()
    plt.figure(figsize=(10, 5))
    plt.plot(global_data["Year"], global_data["No. of Internet Users"] / 1e9, marker="o")
    plt.title("Crescita globale degli utenti Internet (miliardi)")
    plt.xlabel("Anno")
    plt.ylabel("Utenti Internet (miliardi)")
    plt.grid(True)
    plt.show()

# 2. Percentuale media globale di utenti Internet
def plot_percentage_users():
    avg_pct = df.groupby("Year")["Internet Users(%)"].mean().reset_index()
    plt.figure(figsize=(10, 5))
    plt.plot(avg_pct["Year"], avg_pct["Internet Users(%)"], marker="o", color="deepskyblue")
    plt.title("Percentuale media globale di utenti Internet")
    plt.xlabel("Anno")
    plt.ylabel("Percentuale (%)")
    plt.grid(True)
    plt.show()

# 5. Top 10 paesi per numero di utenti Internet (2023)
def plot_top10_users():
    df_latest = df[df["Year"] == df["Year"].max()]
    top10 = df_latest.sort_values("No. of Internet Users", ascending=False).head(10)
    plt.figure(figsize=(10, 6))
    sns.barplot(data=top10, x="No. of Internet Users", y="Entity", palette="viridis")
    plt.title("Top 10 paesi per numero di utenti Internet (2023)")
    plt.xlabel("Utenti Internet")
    plt.ylabel("Paese")
    plt.grid(True)
    plt.show()

# 9. Percentuale utenti Internet per paese (2023)
def plot_pct_per_country():
    df_latest = df[df["Year"] == df["Year"].max()]
    df_top = df_latest[df_latest["Internet Users(%)"] > 0].copy()
    df_top.sort_values("Internet Users(%)", ascending=True, inplace=True)
    norm = plt.Normalize(df_top["Internet Users(%)"].min(), df_top["Internet Users(%)"].max())
    colors = cm.RdYlGn(norm(df_top["Internet Users(%)"]))
    plt.figure(figsize=(10, 18))
    plt.barh(df_top["Entity"], df_top["Internet Users(%)"], color=colors)
    plt.xlabel("% Popolazione con accesso a Internet (2023)")
    plt.title("Percentuale utenti Internet per paese (2023)")
    plt.grid(True)
    plt.show()

# Esecuzione
plot_total_users()
plot_percentage_users()
plot_top10_users()
plot_pct_per_country()
```
